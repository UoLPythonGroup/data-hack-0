The following code is from Jitse, adapted for Python 3 and commented slightly

```python
import pandas, networkx
```

```python
# Function for reading leedsRoads.json into DataFrame
def readLeedsRoadsJSON():
    import json
    jsonData = json.load(open("pythonGeoHack/leedsRoads.geojson"))
    pandasData = pandas.io.json.json_normalize(jsonData['features'])
    # Remove the following columns
    uselessColumns = [ 'geometry.type', 'type', 'properties.points' ]
    pandasData.drop(uselessColumns, axis = 1, inplace = True)
    # Change column titles from "xxx.yyy" to "yyy"
    # cute use of lambda == "inline function"
    pandasData.rename(columns = lambda x: x.split('.')[-1], inplace = True)
    return pandasData

leedsRoads = readLeedsRoadsJSON()
```

```python
# Make a set of (x,y) coords of all the points in the geometry
points = set(tuple(point) for line in leedsRoads['coordinates'] for point in line)
```

```python
graph = networkx.Graph()
graph.add_nodes_from(points)
for ix,row in leedsRoads.iterrows():
    # ix = index of the row
    # row = the row, as a Pandas "series"
    pointsInLine = row['coordinates']
    roadID = int(row['id'])
    for n in range(len(pointsInLine) - 1):
        graph.add_edge(tuple(pointsInLine[n]), tuple(pointsInLine[n+1]), {'roadID': roadID})
```

```python
print(networkx.info(graph))
```

# Finding the 2 core

The "2 core" of a graph is the maximal (induced) subgraph where every vertex has degree at least 2.

```python
core2 = networkx.k_core(graph,2)
```

```python
print(networkx.info(core2))
```

```python
print("Connected components:",networkx.number_connected_components(core2))
sizes = [len(x) for x in networkx.connected_components(core2)]
sizes.sort(reverse=True)
print(sizes)
```

# Remove un-needed internal nodes

Remove all the nodes which have exactly two neighours: but we want to only do this when the neighbours and not themselves linked by an edge.  The current code just kills Python...

```python
alldone = False
while not alldone:
    for n in core2.nodes_iter():
        nbs = list(networkx.all_neighbors(core2,n))
        if len(nbs)==2 and not core2.has_edge(nbs[0],nbs[1]):
                core2.remove_node(n)
                core2.add_edge(nbs[0],nbs[1])
                # So we have ignored attributes: should probably take account of them!
                break # Have to restart
    else:
        # This gets executed if we exit the for loop not through the break command
        alldone = True
```

# Draw the new network

```python
%matplotlib inline
import matplotlib.pyplot as plt
import mpl_toolkits.basemap
import numpy as np
import fiona
import shapely.geometry
from mpl_toolkits.basemap import Basemap
```

```python
f = fiona.open("pythonGeoHack/leedsAll.shp")
leeds = list(f)
```

```python
# For plotting a local area, we can use a similar mercator map
# define the bottom left and top right coordinates in lat/lon.
leeds_map = Basemap(projection='merc',
                    llcrnrlat=53.7,
                    llcrnrlon=-1.75,
                    urcrnrlat=53.95,
                    urcrnrlon=-1.25)
```

```python
from matplotlib.collections import LineCollection

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)

lines = [[leeds_map(lat, lon) for lat, lon in 
              np.asarray(shapely.geometry.shape(x['geometry']))] for x in leeds]

linecoll = LineCollection(lines, linewidths=0.25)

ax.add_collection(linecoll)

leeds_map.drawmapboundary(color='white')
leeds_map
```

```python
from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('complete_map.pdf')
pp.savefig(fig)
pp.close()
```

```python
# Now just draw the edges in core2
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)

lines = [ [leeds_map(lat,lon) for lat, lon in e]
            for e in core2.edges()]

linecoll = LineCollection(lines, linewidths=0.25)

ax.add_collection(linecoll)

leeds_map.drawmapboundary(color='white')
leeds_map
```
